````python
## Code goes here
#importing pandas and re 
import pandas as pd
import re
````

````python
#reading the recweekdata.csv with panda
recweekdata_df = pd.read_csv("recweekdata.csv")
recweekdata_df
````

````python
#reading the studentprofiles.csv with panda
studentprofiles_df = pd.read_csv("studentprofiles.csv")
studentprofiles_df
````

````python
#merging the 2 csv files
merged_df = pd.merge(recweekdata_df, studentprofiles_df)
````

````python
merged_df
````

````python
#cleaning the birthmonth
merged_df.loc[:,"birthmonth"] = merged_df.loc[:,"birthmonth"].str.replace("10|Oct$","October",regex=True)
merged_df.loc[:,"birthmonth"] = merged_df.loc[:,"birthmonth"].str.replace("11|Nov$","November",regex=True)
merged_df.loc[:,"birthmonth"] = merged_df.loc[:,"birthmonth"].str.replace("12|Dec$","December",regex=True)
merged_df.loc[:,"birthmonth"] = merged_df.loc[:,"birthmonth"].str.replace("1|Jan$","January",regex=True)
merged_df.loc[:,"birthmonth"] = merged_df.loc[:,"birthmonth"].str.replace("2|Feb$","February",regex=True)
merged_df.loc[:,"birthmonth"] = merged_df.loc[:,"birthmonth"].str.replace("3|Mar$","March",regex=True)
merged_df.loc[:,"birthmonth"] = merged_df.loc[:,"birthmonth"].str.replace("4|Apr$","April",regex=True)
merged_df.loc[:,"birthmonth"] = merged_df.loc[:,"birthmonth"].str.replace("5|May$","May",regex=True)
merged_df.loc[:,"birthmonth"] = merged_df.loc[:,"birthmonth"].str.replace("6|Jun$","June",regex=True)
merged_df.loc[:,"birthmonth"] = merged_df.loc[:,"birthmonth"].str.replace("7|Jul$","July",regex=True)
merged_df.loc[:,"birthmonth"] = merged_df.loc[:,"birthmonth"].str.replace("8|Aug$","August",regex=True)
merged_df.loc[:,"birthmonth"] = merged_df.loc[:,"birthmonth"].str.replace("9|Sep$","September",regex=True)
````

````python
merged_df
````

````python
#cleaning the birthyear
merged_df.loc[:,"birthyear"] = merged_df.loc[:,"birthyear"].apply(lambda x: re.sub(r"^97","1997", str(x))) 
merged_df.loc[:,"birthyear"] = merged_df.loc[:,"birthyear"].apply(lambda x: re.sub(r"^98","1998", str(x))) 
merged_df.loc[:,"birthyear"] = merged_df.loc[:,"birthyear"].apply(lambda x: re.sub(r"^99","1999", str(x))) 
````

````python
merged_df
````

````python
#cleaning the courses
merged_df.loc[:,"course"] = merged_df.loc[:,"course"].str.replace("^Managment Engineering|BS ME|ME","BS Management Engineering",regex=True)
merged_df.loc[:,"course"] = merged_df.loc[:,"course"].str.replace("^Legal Management|BS LM|LM","BS Legal Management",regex=True)
merged_df.loc[:,"course"] = merged_df.loc[:,"course"].str.replace("^Communications Technology|BS CTM|CTM","BS Communications Technology Management",regex=True)
merged_df.loc[:,"course"] = merged_df.loc[:,"course"].str.replace("^Information Technology Entrepreneurship|BS ITE|ITE","BS Information Technology Entrepreneurship",regex=True)
merged_df.loc[:,"course"] = merged_df.loc[:,"course"].str.replace("^Management|BS MGT|MGT","BS Management",regex=True)
````

````python
merged_df
````

````python
#cleaning the phonenumbers
merged_df.loc[:,"phonenumber"] = merged_df.loc[:,"phonenumber"].str.replace(r'^(9[0-9][0-9])',r'+63 \1',regex=True)
````

````python
merged_df
````

````python
#removing the spaces in between the numbers of the phonenumbers
merged_df.loc[:,"phonenumber"] = merged_df.loc[:,"phonenumber"].apply(lambda x: re.sub(r" ","", str(x)))
````

````python
merged_df
````

````python
#adding a carrier column
carrier = {"+63917":"Globe","+63918":"Smart", "+63927":"Globe"}
merged_df["carrier"] = merged_df.loc[:,"phonenumber"].str[0:6].apply(lambda x: carrier[x])
````
